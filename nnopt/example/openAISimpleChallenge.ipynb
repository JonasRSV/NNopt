Try as many of the easy gyms as possible with one strategy
---

```python
import gym
import numpy as np
from nnopt.nnopt import Optimizer

ITER = 10
```

Cartpole-V1 (Max score 500)
---

```python
cartpole = gym.make("CartPole-v1")
def cartpole_eval(*args, render=False):
    transform = np.array(args).reshape(2, 4)
    
    score = 0
    for _ in range(ITER):
        s = cartpole.reset()
        done = False
        while not done:
            if render:
                cartpole.render()
            s, r, done, _ = cartpole.step(np.argmax(transform @ s))
            score += r
    
    return score / ITER
            
    
```

```python
Nopt = Optimizer(cartpole_eval, N=8, surrogate_hidden_layer=400, Rn=(-1, 1))

domain, score = Nopt.run(random=20, optimization=10, fitting=2000, exploration=0.0, verbose=2)
```

```python
cartpole_eval(*domain, render=True)
```

Space Craft (Anything positive is great)
---

```python
space = gym.make("LunarLanderContinuous-v2")
def space_eval(*args, render=False):
    transform = np.array(args).reshape(2, 8)
    
    score = 0
    for _ in range(ITER):
        s = space.reset()
        done = False
        while not done:
            if render:
                space.render()
            s, r, done, _ = space.step(np.clip(transform @ s, -1, 1))
            score += r
    
    return score / ITER
```

```python
Nopt = Optimizer(space_eval, N=16, surrogate_hidden_layer=400, Rn=(-10, 10), scope="nnopt4")

domain, score = Nopt.run(random=200, optimization=10, fitting=2000, exploration=0.1, verbose=2)
```

```python
space_eval(*domain, render=True)
```
